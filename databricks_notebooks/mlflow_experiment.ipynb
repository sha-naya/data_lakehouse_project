```
import mlflow
import mlflow.sklearn
import pandas as pd
import numpy as np
from numpy import savetxt
import matplotlib.pyplot as plt
 
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```
denorm_basketball_nba = spark.read.table("cs779.denorm_basketball_nba")
```

```
X = np.array(denorm_basketball_nba.select("quarter_1_score").collect())
y = np.array(denorm_basketball_nba.select("total_points_scored").collect())
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=17)
```

```
with mlflow.start_run():
  lr_model = LinearRegression()
  lr_model.fit(X_train, y_train)
  
  predictions = lr_model.predict(X_test)
  
  # Define a metric to use to evaluate the model.
  r2 = r2_score(y_test, predictions)
    
  # Log the value of the metric from this run.
  mlflow.log_metric("r2_score", r2)
  
  # Log the model created by this run.
  mlflow.sklearn.log_model(lr_model, "my_linear_regression_model") 
  
  # Save the table of predicted values
  savetxt('lr_model_predictions.csv', predictions, delimiter=',')
  
  # Log the saved table as an artifact
  mlflow.log_artifact("lr_model_predictions.csv")
```
